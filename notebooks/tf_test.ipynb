```python
# Now trying on our own data
import os
import glob
import numpy as np
from skimage.io import imsave, imread
from skimage.transform import resize

img_cols = 1392
img_rows = 1040
target_width = 696 #1392
target_height = 520 #1040

patch_size = 256

def read_imgs_masks2(img_dir, mask_dir, img_rows, img_cols,
                    target_width=target_width, target_height=target_height):
    images = [f for f in os.listdir(img_dir) if f.endswith('.tif')]
    masks = [f for f in os.listdir(mask_dir) if f.endswith('.tif')]
    
    if len(images)!=len(masks):
        print("Images and mask folder have unequal number of images.\nOnly files existing in both directories will be returned.")
    
    m  = np.min([len(images),len(masks)],axis=0)
    imgs = np.ndarray((int(m), target_height, target_width, 1), dtype=np.float)
    masks = np.ndarray((int(m), target_height, target_width, 1), dtype=np.float)
    c=0
    for idx, img in enumerate(images):
        #print(idx)
        try:
            msk = img.replace('.tif', '_Simple Segmentation.tif')
            img = imread(os.path.join(img_dir, img), as_gray=True)
            img = resize(img, (target_height, target_width))
            img = np.expand_dims(img, axis=-1)
            imgs[c] = img

            mask = imread(os.path.join(mask_dir, msk), as_gray=True)
            mask = resize(mask, (target_height, target_width))
            mask = np.expand_dims(mask, axis=-1)
            masks[c] = mask
            c+=1
        except:
            continue
    return imgs, masks
```

```python
test_image_tf = dict(samplewise_center=False,
                    samplewise_std_normalization=False,
                    rescale=True,
                    vertical_flip=False,
                    horizontal_flip=False,
                    preprocessing_function=False,
                    fill_mode=True,
                    max_iter = False,
                    thresh_obj_perc = False,
                    target_width = False,
                    target_height = False
                    )
    
test_mask_tf = dict(samplewise_center=False,
                samplewise_std_normalization=False,
                rescale=False,
                vertical_flip=False,
                horizontal_flip=False,
                preprocessing_function=True,
                fill_mode=True,
                max_iter = False,
                thresh_obj_perc = False,
                target_width = False,
                target_height = False
                )

img_mask_args_avg_test_per = dict(samplewise_center=False,
                    samplewise_std_normalization=False,
                    rescale=1./255,
                    vertical_flip=False,
                    horizontal_flip=False,
                    preprocessing_function=np.rint,
                    fill_mode='nearest',
                    max_iter = None,
                    thresh_obj_perc = 0.99,
                    target_width = img_cols,
                    target_height = img_rows,
                    image_aug_check = test_image_tf,
                    mask_aug_check = test_mask_tf
                     )

image_mask_datagen_avg_test_per = customImageDataGenerator(**img_mask_args_avg_test_per)


test_imgs_path = '../../ds1/validation/img/cls'
test_masks_path = '../../ds1/validation/man_mask/cls'
```

```python
def divide_img_blocks(img, n_blocks=(1280//256, 1536//256)):
    horizontal = np.array_split(img, n_blocks[0])
    # print(len(horizontal))
    splitted_img = [np.array_split(block, n_blocks[1], axis=1) for block in horizontal]
    # print(len(splitted_img))
    return np.asarray(splitted_img, dtype=np.ndarray)
```

```python
all_images = np.empty(shape=((((target_height//patch_size+1)*patch_size)//patch_size)*
            (((target_width//patch_size+1)*patch_size)//patch_size),patch_size,patch_size,1))

all_masks = np.empty(shape=((((target_height//patch_size+1)*patch_size)//patch_size)*
(((target_width//patch_size+1)*patch_size)//patch_size),patch_size,patch_size,1))

for f in glob.glob(test_imgs_path+'/*.tif'):
#     print(f)
    img_path = f
    msk_path = test_masks_path+'/'+f.split('/')[-1].replace('.tif','_Simple Segmentation.tif')
 
    img = load_img(
            img_path,
            color_mode="grayscale",
            target_size=(target_height,target_width),
            interpolation="nearest",
            keep_aspect_ratio=True,
        )
    mask = load_img(
            msk_path,
            color_mode="grayscale",
            target_size=(target_height,target_width),
            interpolation="nearest",
            keep_aspect_ratio=True,
            )
    # print('objects read')
    x = img_to_array(img, data_format='channels_last')
    y = img_to_array(mask, data_format='channels_last')
    # Pillow images should be closed after `load_img`,
    # but not PIL images.
    if hasattr(img, "close"):
        img.close()
    if hasattr(mask, "close"):
        mask.close() 

    x = image_mask_datagen_avg_test_per.standardize(x,test_image_tf)
    y = image_mask_datagen_avg_test_per.standardize(y,test_mask_tf)
    # print('standardized')

    # print(x.shape)
    # print(y.shape)

    x_padded = np.pad(np.squeeze(x), ((0, ((target_height//patch_size+1)*patch_size)-target_height), 
                                      (0, ((target_width//patch_size+1)*patch_size)-target_width)),
                        mode='constant', constant_values=0)
    y_padded = np.pad(np.squeeze(y), ((0, ((target_height//patch_size+1)*patch_size)-target_height), 
                                      (0, ((target_width//patch_size+1)*patch_size)-target_width)),
                        mode='constant', constant_values=0)
                      
    # print(x_padded.shape)

    img_array = divide_img_blocks(np.expand_dims(x_padded,axis=-1), 
              n_blocks=(((target_height//patch_size+1)*patch_size)//patch_size, 
                        ((target_width//patch_size+1)*patch_size)//patch_size))
                      
    msk_array = divide_img_blocks(np.expand_dims(y_padded,axis=-1), 
              n_blocks=(((target_height//patch_size+1)*patch_size)//patch_size, 
                        ((target_width//patch_size+1)*patch_size)//patch_size))
    
    img_array2  = np.zeros(shape=(img_array.shape[0]*img_array.shape[1], img_array.shape[2],img_array.shape[3],
                                 img_array.shape[4]))
    msk_array2  = np.zeros(shape=(msk_array.shape[0]*msk_array.shape[1], msk_array.shape[2],msk_array.shape[3],
                                 msk_array.shape[4]))
    
    c=0
    for i in range(img_array.shape[0]):
        for j in range(img_array.shape[1]):
            img_array2[c] = img_array[i,j]
            c+=1
            
    c=0
    for i in range(msk_array.shape[0]):
        for j in range(msk_array.shape[1]):
            msk_array2[c] = msk_array[i,j]
            c+=1
    all_images = np.vstack((all_images,img_array2))
    all_masks = np.vstack((all_masks,msk_array2))      
#     for i in range(img_array2.shape[0]):
#         plt.figure()
#         plt.imshow(img_array2[i],cmap='gray')
#     break
    
        
```
